```python
df = pd.read_csv('https://raw.githubusercontent.com/saujanya01/kprototype_clustering/master/with_cluster_full.csv')
df.head(10)
```

## Categorical Data

```python
ind = []
l = df.dtypes
for i in range(len(l)):
    if (str(l[i])=='object'):
        ind.append(i)
c_data = [df.columns[i] for i in ind]
c_data
```

# Numerical Data

```python
plt.figure(figsize=(15,10))
sns.set(font_scale=1.25)
sns.kdeplot(df.loc[(df['cluster']==0),'age'], color='r', Label='Cluster 1')
sns.kdeplot(df.loc[(df['cluster']==1),'age'], color='b', Label='Cluster 2')
sns.kdeplot(df.loc[(df['cluster']==2),'age'], color='g', Label='Cluster 3')
sns.kdeplot(df.loc[(df['cluster']==3),'age'], color='y', Label='Cluster 4')
plt.xlabel("Age")
plt.ylabel("PDF")
# plt.setp(axes,yticks=[])
# plt.tight_layout()
```

```python
f, axes = plt.subplots(2, 2, figsize=(15, 10), sharex=True)
sns.set(font_scale=1.25)
sns.kdeplot(df.loc[(df['cluster']==0),'campaign_contact_count'], color='r', Label='Cluster 1', ax = axes[0,0])
plt.xlabel("campaign contact count")
sns.kdeplot(df.loc[(df['cluster']==1),'campaign_contact_count'], color='b', Label='Cluster 2', ax = axes[0,1])
plt.xlabel("campaign contact count")
sns.kdeplot(df.loc[(df['cluster']==2),'campaign_contact_count'], color='g', Label='Cluster 3', ax = axes[1,0])
plt.xlabel("campaign contact count")
sns.kdeplot(df.loc[(df['cluster']==3),'campaign_contact_count'], color='y', Label='Cluster 4', ax = axes[1,1])
plt.xlabel("campaign contact count")
plt.ylabel("PDF")
# plt.setp(axes)
plt.tight_layout()
```

```python
f, axes = plt.subplots(2, 2, figsize=(15, 10), sharex=True)
sns.set(font_scale=1.25)
sns.kdeplot(df.loc[(df['cluster']==0),'last_contact_duration'], color='r', Label='Cluster 1', ax = axes[0,0])
plt.xlabel("campaign contact count")
sns.kdeplot(df.loc[(df['cluster']==1),'last_contact_duration'], color='b', Label='Cluster 2', ax = axes[0,1])
plt.xlabel("campaign contact count")
sns.kdeplot(df.loc[(df['cluster']==2),'last_contact_duration'], color='g', Label='Cluster 3', ax = axes[1,0])
plt.xlabel("campaign contact count")
sns.kdeplot(df.loc[(df['cluster']==3),'last_contact_duration'], color='y', Label='Cluster 4', ax = axes[1,1])
plt.xlabel("last_contact_duration in seconds")
plt.ylabel("PDF")
# plt.setp(axes)
plt.tight_layout()
```
