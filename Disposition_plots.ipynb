```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
st = '''No Answer-780000 customer
SIT Invalid Number-140700 process
Stale-39000 process
Abandoned-37000 system
Dropped-36000 customer
Ringing No response-31000 customer
Customer Disconnecting the call-30000 customer
Not Reachable-24000 customer
SIT NC (No Circuit)-12000 system
Busy-10000 customer
Phone Busy-7200 customer
Remote Release-4300 system
Dial Error-2800 system
does not exist-1000 process'''
```

```python
st
```

```python
l=st.split("\n")
```

```python
li = [[i.split('-')[0],i.split('-')[1].split(' ')[0],i.split('-')[1].split(' ')[1]] for i in l]
```

```python
li[1:]
```

```python
df = pd.DataFrame(li[1:],columns=['disposition','count','category'])
```

```python
df[['count']]=df[['count']].apply(pd.to_numeric)
```

```python
sns.set(font_scale=1.25)
plt.figure(figsize=(15,10))
sns.barplot(x='count',y='disposition',data=df,hue='category')
```

```python
total = df['count'].sum()
```

```python
total
```

```python
process = df[df['category']=='process']['count'].sum()
customer = df[df['category']=='customer']['count'].sum()
system = df[df['category']=='system']['count'].sum()
print(process)
print(customer)
print(system)
```
